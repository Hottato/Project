```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Define the structure of the network with English labels
layers = [
    ('Current Prediction Data', 'Hidden Node1'), 
    ('Current Prediction Data', 'Hidden Node2'),
    ('Current Prediction Data', 'Hidden Node3'),
    ('Current Prediction Data', 'Hidden Node4'),
    ('Wear Pixel Count', 'Hidden Node1'), 
    ('Wear Pixel Count', 'Hidden Node2'),
    ('Wear Pixel Count', 'Hidden Node3'),
    ('Wear Pixel Count', 'Hidden Node4'),
    ('Node3', 'Hidden Node1'), 
    ('Node3', 'Hidden Node2'),
    ('Node3', 'Hidden Node3'),
    ('Node3', 'Hidden Node4'),
    ('Hidden Node1', '?'), 
    ('Hidden Node2', '?'),
    ('Hidden Node3', '?'),
    ('Hidden Node4', '?'),
]

# Create the graph
G = nx.DiGraph()
G.add_edges_from(layers)

# Define the positions for the nodes
pos = {
    'Current Prediction Data': [0, 2], 'Wear Pixel Count': [0, 1], 'Node3': [0, 0],
    'Hidden Node1': [1, 2.5], 'Hidden Node2': [1, 1.5], 'Hidden Node3': [1, 0.5], 'Hidden Node4': [1, -0.5],
    '?': [2, 1]
}

# Draw the graph
plt.figure(figsize=(20, 8))  # Increase the figure size
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=10000, font_size=18, font_weight='bold', arrows=True)
plt.title('Neural Network Structure', fontsize=20)
plt.show()
```
