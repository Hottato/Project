```python
import cv2
import numpy as np
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
from PIL import Image

blurred_image_path = r"C:\Users\31083\jojojojo\Project2\CNC_YOLOV8_Segmodel\Seminar_Presentation\blurred_image (1).jpg"  # replace with your image path
blurred_img_pil = cv2.imread(blurred_image_path, cv2.IMREAD_GRAYSCALE)

equalized_img_path = r"C:\Users\31083\jojojojo\Project2\CNC_YOLOV8_Segmodel\Seminar_Presentation\clahe_image (2).jpg" 
clahe_img = cv2.imread(equalized_img_path, cv2.IMREAD_GRAYSCALE)

blurred_histogram = np.histogram(blurred_img_pil, bins=256, range=(0,256))[0]
clahe_histogram = np.histogram(clahe_img, bins=256, range=(0,256))[0]


aa_path = r"C:\Users\31083\jojojojo\Project2\CNC_YOLOV8_Segmodel\Seminar_Presentation\aaa.png"  # replace with your image path
aa = cv2.imread(aa_path, cv2.IMREAD_GRAYSCALE)

bb_path = r"C:\Users\31083\jojojojo\Project2\CNC_YOLOV8_Segmodel\Seminar_Presentation\bbb.png" 
bb = cv2.imread(bb_path, cv2.IMREAD_GRAYSCALE)


be_path = r"C:\Users\31083\jojojojo\Project2\CNC_YOLOV8_Segmodel\Seminar_Presentation\be.png" 
be = cv2.imread(be_path, cv2.IMREAD_GRAYSCALE)
```

```python



# Define font size
font_size = 30 # change to your preferred font size

# Display images and histograms
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(30, 20))  # change figsize to adjust image size

axes[0, 0].imshow(blurred_img, cmap='gray')
axes[0, 0].set_title('Blurred Image', fontsize=font_size)
axes[0, 0].axis('off')

axes[0, 1].bar(range(256), blurred_histogram, color='gray', alpha=0.7)
axes[0, 1].set_title("Histogram of the Blurred Image", fontsize=font_size)
axes[0, 1].set_xlabel("Pixel Value", fontsize=font_size)
axes[0, 1].set_ylabel("Frequency", fontsize=font_size)

axes[1, 0].imshow(clahe_img, cmap='gray')
axes[1, 0].set_title('CLAHE Image', fontsize=font_size)
axes[1, 0].axis('off')

axes[1, 1].bar(range(256), clahe_histogram, color='gray', alpha=0.7)
axes[1, 1].set_title("Histogram of the CLAHE Image", fontsize=font_size)
axes[1, 1].set_xlabel("Pixel Value", fontsize=font_size)
axes[1, 1].set_ylabel("Frequency", fontsize=font_size)


plt.tight_layout()
plt.subplots_adjust(hspace=0.2) 
plt.show()
```
