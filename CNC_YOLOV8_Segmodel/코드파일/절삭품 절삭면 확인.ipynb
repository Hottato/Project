# import

```python
import numpy as np, cv2
import matplotlib.pylab as plt
import os
```

```python
# 폴더 경로
nor_folder_path = "C:/Users/31083/Documents/GitHub/Project/CNC_YOLOV8_Segmodel/normal_maybe"

image_files = os.listdir(nor_folder_path)

nor_images = []
for file_name in image_files:
    file_path = os.path.join(nor_folder_path, file_name)
    image = cv2.imread(file_path, cv2.IMREAD_COLOR)

    nor_images.append(image)
    
    
# 폴더 경로
wear_folder_path = "C:/Users/31083/Documents/GitHub/Project/CNC_YOLOV8_Segmodel/waer_maybe"

image_files = os.listdir(wear_folder_path)

wear_images = []
for file_name in image_files:
    file_path = os.path.join(wear_folder_path, file_name)
    image = cv2.imread(file_path, cv2.IMREAD_COLOR)

    wear_images.append(image)
    
```

# Edge detection

`-` 정상이미지( 가우시안-> Edge detection)

```python
blurred = cv2.GaussianBlur(wear_images[1], (5, 5), 0)

plt.imshow(255-blurred)

canny = cv2.Canny(blurred, 100,300)

plt.subplot(1,2,1)
plt.title('wear')
plt.axis('off')
plt.imshow(wear_images[1])

plt.subplot(1,2,2)
plt.title('Edge Detection')
plt.axis('off')
plt.imshow(canny, cmap='gray')

plt.show()
```

`-` 마모 이미지( 가우시안-> Edge detection)

```python

blurred = cv2.GaussianBlur(nor_images[2], (5, 5), 0)
#흑백변환


canny = cv2.Canny(255-blurred, 100,300)

plt.imshow(canny)

plt.subplot(1,2,1)
plt.title('nor')
plt.axis('off')

plt.imshow(nor_images[2], cmap='gray')

plt.subplot(1,2,2)
plt.title('Edge Detection')
plt.axis('off')
plt.imshow(canny, cmap='gray')

plt.show()
```

> 두 이미지의 차이가 파악을 못하겠음
