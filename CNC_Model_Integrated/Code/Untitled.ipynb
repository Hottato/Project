```python
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt

import plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
# 데이터 파일 경로
data = 'C:\\Users\\31083\\Documents\\GitHub\\Project\\CNC_Model_Integrated\\Data'
data
```

```python
one  = pd.read_csv(f"{data}\\2022.12.02(14번 - feed 120)\\1번째 데이터.csv",skiprows=1)
last = pd.read_csv(f"{data}\\2022.12.02(14번 - feed 120)\\14번째 데이터.csv",skiprows=1)
one.columns
```

```python
one = one[one[' z'] <0]

last = last[last[' z'] <0]
```

```python

# Assuming 'one' and 'last' DataFrames contain the data for the first and last traces

fig = make_subplots(rows=1, cols=1)

fig.add_trace(
    go.Scatter(x=one['hh:mm:ss:ms'], y=one[' C1'], mode='lines', name='one_C1')
)

fig.add_trace(
    go.Scatter(x=one['hh:mm:ss:ms'], y=last[' C1'], mode='lines', name='last_C1')
)

fig.update_yaxes(range=[0, 10])
fig.show()
```
